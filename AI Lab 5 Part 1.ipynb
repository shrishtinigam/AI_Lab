```python
import pandas as pd          
from sklearn.naive_bayes import GaussianNB
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import metrics

dataset = pd.read_csv('heart.csv')
dataset.head(5)

# Changing the string values into binary encoded form
replacement = pd.get_dummies(dataset.Sex)
replaced = pd.concat([dataset, replacement], axis = 1)
dataset = replaced.drop(['Sex', 'F'], axis = 1)


replacement = pd.get_dummies(dataset.ChestPainType)
replaced = pd.concat([dataset, replacement], axis = 1)
dataset = replaced.drop(['ChestPainType', 'M'], axis = 1)

replacement = pd.get_dummies(dataset.RestingECG)
replaced = pd.concat([dataset, replacement], axis = 1)
dataset = replaced.drop(['RestingECG', 'Normal'], axis = 1)


replacement = pd.get_dummies(dataset.ExerciseAngina)
replaced = pd.concat([dataset, replacement], axis = 1)
dataset = replaced.drop(['ExerciseAngina', 'N'], axis = 1)


replacement = pd.get_dummies(dataset.ST_Slope)
replaced = pd.concat([dataset, replacement], axis = 1)
dataset = replaced.drop(['ST_Slope', 'Flat'], axis = 1)

print(dataset.head(5))

X = dataset.drop(columns = 'HeartDisease', axis = 1) # for rows axis = 0 should be specified
Y = dataset['HeartDisease']

# Scaling data to decrease outliers
scaler = StandardScaler()
standardized_data = scaler.fit_transform(X)

# Splitting data
X_Train, X_Test, Y_Train, Y_Test = train_test_split(X, Y, test_size = 0.2, stratify = Y, random_state = 3)

model  =  GaussianNB()
model.fit(X_Train, Y_Train) 

prediction  =  model.predict(X_Test)

from sklearn.metrics import confusion_matrix
print("\nConfusion Matrix")
print(confusion_matrix(Y_Test, prediction))

accuracy  =  model.score(X_Test, Y_Test)
print("Naive Bayes accuracy is :", accuracy)
```
