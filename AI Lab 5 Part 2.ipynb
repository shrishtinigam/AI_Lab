```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB

dataset = pd.read_csv('tennis (2).csv')

print('No of rows:', dataset.shape[0])
print('nO of columns:', dataset.shape[1])

print(dataset.head(10))

dataset.info()

# Class prior
no_of_plays  =  dataset.groupby('play').size()
yes_play = no_of_plays['Yes']
no_play = no_of_plays['No']
ypc = yes_play/dataset.shape[0] # PROBABILITY OF YES
ncp = no_play/dataset.shape[0] # PROBABILITY OF NO
print("Yes class probability")
print(ypc)
print("No class probability")
print(ncp)

tempplay  =  dataset.groupby(['temp',  'play']).size()
print("temp_play\n", tempplay)

humidityplay  =  dataset.groupby(['humidity',  'play']).size()
print("humid_play\n", humidityplay)

windplay  =  dataset.groupby(['wind',  'play']).size()
print("wind_play\n", windplay)

POY  =  tempplay['Overcast',  'Yes']
PSY  =  tempplay['Sunny',  'Yes']
PRY  =  tempplay['Rain',  'Yes']
PSN  =  tempplay['Sunny',  'No']
PRN  =  tempplay['Rain',  'No']

PCY  =  humidityplay['Cool',  'Yes']
PHY  =  humidityplay['Hot',  'Yes']
PMY  =  humidityplay['Mild',  'Yes']
PCN  =  humidityplay['Cool',  'No']
PHN  =  humidityplay['Hot',  'No']
PMN  =  humidityplay['Mild',  'No']

PHIY  =  windplay['High',  'Yes']
PNY  =  windplay['Normal',  'Yes']
PHIN =  windplay['High',  'No']
PNN  =  windplay['Normal',  'No']

# testing <Rain Cool High>
LY = (PRY/yes_play) *(PCY/yes_play)*(PHIY/yes_play)
print('likelihood:Yes', LY)
LN = (PRY/no_play) *(PCY/no_play)*(PHIY/no_play)
print('Likelihood:No', LN)

CY = LY*ypc
CN = LN*ncp
print('Class Yes:', CY)
print('Class No:', CN)

dataset = dataset.drop(['outlook'], axis = 1)
print(dataset)

replacement = pd.get_dummies(dataset)
print(replacement)
print(dataset)

dataset = pd.concat([dataset, replacement], axis = 1)
print(dataset)

dataset = dataset.drop(['temp', 'humidity', 'wind', 'play', 'temp_Overcast','humidity_Cool','wind_High','play_No'], axis = 1)

print(dataset.head(5))

X = dataset.drop(['play_Yes'], axis = 1)
Y = dataset['play_Yes']

# Splitting into train and test data
X_train,  X_test,  Y_train,  Y_test  =  train_test_split(X, Y, test_size = 0.2, random_state = 1)

model  =  GaussianNB()
model.fit(X_train, Y_train) 
accuracy  =  model.score(X_test, Y_test)
print("Accuracy:", accuracy)
print('Class count:', model.class_count_)
print('Class prior', model.class_prior_)

prediction =  model.predict([[1, 0, 0, 0, 0]])
print (prediction)
```
